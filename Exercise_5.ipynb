```python

"""A template for a python script deliverable for INST326.
Driver: Zain Bhatti
Navigator: Alejandro Tarazona
Assignment: exercise 5
Date: 04_04_22
"""

"""A program that parses though a txt file using regular expressions then separates wanted information from the txt file"""
 
import re
class Address:
    """An Address class that acts as a data container
        
    Attributes:
        street (String): a string indicating the street
        city (String): a sting indicating the city
        state (String):  a string indicating the state
    """
    
    def __init__(self, street, city, state):
        """Initializes the Address objects
        
        Args: 
            street (String): see class documentation
            city (String): see class documentation
            state (String): see class documentation
        """
        
        self.street = street
        self.city = city
        self.state = state
        
class Employee:
    """An Employee class that acts as a data container
        
    Attributes:
        text (String): The text file with all of the data we want to parse through
    """
    
    def __init__(self, text):
        """Initializes the Employee objects
        
        Args: 
            text (String): see class documentation
        """
        
        name = parse_name(text)
        self.first_name = name[0]
        self.last_name = name[1]
        self.address = parse_address(text)
        self.email = parse_email(text)

def parse_name(text):
    """Function uses regular expressions in order to capture the first name and last
        name of the person in question
        
        Args:
            text (String): The text file with all of the data we want to parse through
            
        Returns:
            fNamelName (Tuple):  Tuple that contains the first and last name as strings
        """
    
    firstName = str(re.findall("[A-Za-z]", text))
    lastName = str(re.findall("[A-Za-z]", text))
    
    fNamelName = (str(firstName), str(lastName))
    return fNamelName

def parse_address(text):
    """Function uses regular expressions in order to capture the street, city, and state
        of the person in question
        
        Args:
            text (String): The text file with all of the data we want to parse through
            
        Returns:
            addy (Object): An address object using the street, city and state identified
        """
    
    addy = str(re.findall('[A-Za-z0-9]', text))
    
    return addy

def parse_email(text):
    """Function uses regular expressions in order to capture the email
        of the person in question
        
        Args:
            text (String): The text file with all of the data we want to parse through
            
        Returns:
            mail (String): The email identified
        """
    
    mail = str(re.findall('\w+.\w+.\w'), text)
    
    return mail

def main(path):
    """The main function that opens and reads the txt file while calling other methods
        
        Args:
            path (String): The text file with all of the data we want to parse through
        """
    
    employee_list = []
    with open(path, "r") as file:
        
        for line in file:
            
            emp = Employee(line)
            employee_list.append(emp)
            
    file.close()
         
if __name__ == "__main__":
    
    main("people.txt")
```
